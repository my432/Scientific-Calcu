```python
import math

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Cannot divide by zero!"

def power(x, y):
    return x ** y

def square_root(x):
    return math.sqrt(x)

def logarithm(x, base=10):
    return math.log(x, base)

def exp(x):
    return math.exp(x)

def sine(x):
    return math.sin(math.radians(x))

def cosine(x):
    return math.cos(math.radians(x))

def tangent(x):
    return math.tan(math.radians(x))

def factorial(x):
    return math.factorial(x)

def scientific_calculator():
    print("Scientific Calculator")
    print("----------------------")
    print("Select operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Power")
    print("6. Square Root")
    print("7. Logarithm")
    print("8. Exponential (e^x)")
    print("9. Sine")
    print("10. Cosine")
    print("11. Tangent")
    print("12. Factorial")

    choice = input("Enter choice (1-12): ")

    if choice in ['1', '2', '3', '4', '5']:
        x = float(input("Enter first number: "))
        y = float(input("Enter second number: "))

        if choice == '1':
            print(f"{x} + {y} = {add(x, y)}")
        elif choice == '2':
            print(f"{x} - {y} = {subtract(x, y)}")
        elif choice == '3':
            print(f"{x} * {y} = {multiply(x, y)}")
        elif choice == '4':
            print(f"{x} / {y} = {divide(x, y)}")
        elif choice == '5':
            print(f"{x} ^ {y} = {power(x, y)}")

    elif choice == '6':
        x = float(input("Enter number: "))
        print(f"Square root of {x} = {square_root(x)}")

    elif choice == '7':
        x = float(input("Enter number: "))
        base = input("Enter base (default is 10): ")
        base = float(base) if base else 10
        print(f"log({x}, base {base}) = {logarithm(x, base)}")

    elif choice == '8':
        x = float(input("Enter number: "))
        print(f"e^{x} = {exp(x)}")

    elif choice in ['9', '10', '11']:
        x = float(input("Enter angle in degrees: "))

        if choice == '9':
            print(f"sin({x}) = {sine(x)}")
        elif choice == '10':
            print(f"cos({x}) = {cosine(x)}")
        elif choice == '11':
            print(f"tan({x}) = {tangent(x)}")

    elif choice == '12':
        x = int(input("Enter a non-negative integer: "))
        print(f"{x}! = {factorial(x)}")

    else:
        print("Invalid choice. Please try again.")

# Run the calculator
scientific_calculator()
```
